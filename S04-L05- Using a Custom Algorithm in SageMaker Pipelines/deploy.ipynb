```python
import boto3
import sagemaker
import time

session = sagemaker.Session()
bucket = session.default_bucket()
role = sagemaker.get_execution_role()
```

```python
from sagemaker import ModelPackage
from sagemaker.serializers import CSVSerializer

model_package_arn = 'arn:aws:sagemaker:eu-west-1:913036176028:model-package/churn-model-group/91'
model = ModelPackage(
    role=role,
    model_package_arn=model_package_arn,
    sagemaker_session=session
)

endpoint_name = 'churn-model-endpoint-2'
model.deploy(
    initial_instance_count=1,
    instance_type='ml.m5.large',
    endpoint_name=endpoint_name,
)
```

```python
from sagemaker.predictor import Predictor

predictor = Predictor(
    endpoint_name=endpoint_name,
    serializer=CSVSerializer()
)
```

```python
import pandas as pd

churn_df_test = pd.read_csv('./churn_test.csv')
```

```python
churn_np_test = churn_df_test.to_numpy()
```
